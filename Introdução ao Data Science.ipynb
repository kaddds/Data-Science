# O que é Data Science?

O trabalho de cientista de dados já foi chamado de "o emprego mais sexy do século XXI".

https://hbr.org/2012/10/data-scientist-the-sexiest-job-of-the-21st-century

💡**Realidade!!!** 

Ciência de Dados é uma disciplina que combina estatística, matemática, ciência da computação e conhecimentos de domínio para **extrair insights e valor de dados**. A razão pela qual o campo de Data Science se tornou tão popular e essencial em nossa sociedade é simples: **vivemos na era da informação**.

Com a explosão de dados gerados diariamente por empresas, governos e até mesmo por indivíduos em redes sociais, há uma necessidade crescente de entender, analisar e usar essas informações de forma eficaz. 

Aqui estão algumas razões pelas quais Data Science é tão relevante:

**1. Tomada de decisões baseada em dados:** Data Science permite que as empresas usem estatísticas e modelos preditivos para tomar decisões mais informadas e eficazes.
Eficiência operacional: Com a análise de dados, as empresas podem otimizar processos, automatizar tarefas e reduzir custos.

**2. Personalização de serviços:** A Ciência de Dados ajuda a entender o comportamento do cliente, permitindo criar experiências personalizadas e melhorar o engajamento.

**3. Identificação de tendências e oportunidades:** A análise de dados permite que as empresas identifiquem novas oportunidades de mercado e acompanhem as tendências do setor.

**4. Melhoria da segurança e prevenção de fraudes:** Data Science pode ser usada para detectar padrões suspeitos e aprimorar as medidas de segurança.

Data Science oferece um caminho para transformar dados brutos em conhecimento acionável, beneficiando negócios, comunidades e a sociedade como um todo. 

**É uma ferramenta poderosa para lidar com a complexidade do mundo moderno e para impulsionar avanços em diversas áreas.**

# Papel do Data Science nas Empresas

🏢 O papel do Data Science nas empresas é fundamental para **alavancar dados e tomar decisões mais estratégicas e informadas.**

# Exemplos de Uso do Data Science nas Empresas

## Caso de Uso 1: Logística

A ciência de dados tornou-se uma ferramenta poderosa para as empresas de **logística otimizarem suas operações**. Nomeadamente, pode ser utilizado para determinar as **rotas de entrega mais eficientes, gerir a utilização de combustível nas horas do dia mais económicas e fazer previsões mais precisas sobre a oferta e a procura**. E todos esses fatores ajudam as empresas de logística a economizar tempo e dinheiro.

A **UPS** desenvolveu ferramentas de planejamento de rede chamado Network Planning Tools (NPT) para visualizar de forma abrangente as operações da organização. 

Essas ferramentas foram projetadas para fornecer uma visão geral das atividades nas instalações da UPS em todo o mundo e do volume e distribuição de pacotes em sua rede de coleta e entrega. Também permite que a UPS obtenha informações detalhadas sobre remessas em trânsito, como peso, volume e prazos de entrega.

A NPT utiliza aprendizado de máquina para **processar e analisar grandes quantidades de dados do mundo real, permitindo que a UPS visualize todas as suas instalações, desvie seus pacotes durante tempestades ou mova rapidamente uma grande remessa.**

Assim como as melhores ferramentas de ciência de dados, o NPT capacita os humanos a tomar melhores decisões. Quando um pacote é redirecionado, o aplicativo notifica um engenheiro no novo local sobre o plano revisado. O engenheiro então avalia várias opções e toma medidas, potencialmente redirecionando o pacote com base em novas informações. 
O NPT também verifica a escolha do engenheiro para garantir que obteve o resultado desejado. Isso permite que o algoritmo do NPT melhore e tome decisões mais inteligentes quando

Os motoristas da empresa fazem, em média, 120 entregas por dia.

**Ao cruzar os dados dos sensores nos veículos, requisitos das encomendas dos clientes e dados de mapas e geolocalização, a UPS reduziu em 85 milhões o número de milhas percorridas por ano.**

Além do aumento de qualidade de vida para os funcionários, a medida proporcionou uma economia expressiva de combustível.

## Caso de Uso 2: Bancos


A ciência de dados revolucionou a forma como as empresas **detectam atividades fraudulentas**. Ele desempenha um papel crucial na coleta, resumo e previsão de bancos de dados de clientes para identificar atividades fraudulentas com precisão.

Inteligência de Contrato (COiN)
O JP Morgan também usa aprendizado de máquina e inteligência artificial para desenvolver um novo software chamado COIN (Contract Intelligence). A ferramenta **revisou milhares de contratos de crédito do banco**, utilizando reconhecimento de imagem para reconhecer padrões nesses contratos. Esta tecnologia emprega aprendizagem não supervisionada, permitindo ao algoritmo classificar cláusulas em um dos cerca de 150 “atributos” diferentes de contratos de crédito. Por exemplo, o algoritmo pode observar padrões específicos com base no texto da cláusula ou na localização do acordo.

O investimento do JPMorgan em ciência de dados é motivado pelos benefícios da **redução de custos**. O software pode revisar contratos em segundos, um processo que anteriormente exigia mais de 360 mil horas de trabalho dos advogados. O algoritmo não apenas economiza tempo e dinheiro. Também melhora a precisão do processo. Em comparação com advogados humanos, a precisão do COIN é significativamente maior. Assim, o investimento do banco não se resume apenas à redução de custos, mas também ao **aumento da qualidade**.

## Caso de Uso 3: Grupo Pão de Açúcar

A rede varejista brasileira Pão de Açúcar disponibiliza programas de recompensas para os clientes dos seus mercados — Cliente Mais para os mercados Pão de Açúcar e Clube Extra para os mercados Extra. Dessa forma, os clientes acumulam pontos ao fazerem compras e podem trocá-los por descontos em compras futuras.

Com esse sistema, que também está integrado com os fornecedores, o Pão de Açúcar consegue descobrir os produtos favoritos dos consumidores e gerar cupons com ofertas personalizadas.

Além de contribuir para a fidelização dos clientes, o sistema dá à empresa e aos fornecedores dados precisos sobre os produtos mais comprados, favorecendo principalmente as operações de estoque e logística.

## Caso de Uso 4: IBM

No caso da IBM, o desafio era melhorar o clima organizacional. Para isso, a empresa passou a monitorar padrões nas publicações dos colaboradores em redes sociais internas, buscando entender as principais causas de insatisfação.

O segredo foi a utilização da inteligência artificial do IBM Watson, o Social Pulse. Nela, a ideia é, basicamente, buscar palavras-chave e, assim, entender tendências— que representam a voz dos colaboradores. Logo após, o programa cria dashboards com os sentimentos mais identificados, divisão por segmentos, linha do tempo, entre outros.

Em sua aplicação interna, foi possível identificar as insatisfações das equipes, como erros na forma de conceder feedback e na comunicação interna.

## Caso de Uso 5: Itaú

Existem diversas áreas onde podem ser aplicadas soluções de data science.

Em especial na operação de atendimento ao cliente, o Itaú fez o bom uso do data science por utilizar metolodologias de NLP (Natural language processing) durante os atendimentos com os clientes. 

Foi possível identificar, as polaridades dos clientes para saber se a ligação é positiva ou negativa e como a operação pode melhorar o seu atendimento para ter um resultado positivo e satisfatório. 

Além da melhoria no atendimento da operação, é possível prever possíveis problemas que podem ser tratados de forma tempestiva com atencedência e até idenficar qual é a real necessidade do cliente por meio de correlações feitas durante as ligações.
