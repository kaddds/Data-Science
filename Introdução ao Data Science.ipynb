## Caso de Uso 1: Logstica

A cincia de dados tornou-se uma ferramenta poderosa para as empresas de **logstica otimizarem suas operaes**. Nomeadamente, pode ser utilizado para determinar as **rotas de entrega mais eficientes, gerir a utilizao de combustvel nas horas do dia mais econmicas e fazer previses mais precisas sobre a oferta e a procura**. E todos esses fatores ajudam as empresas de logstica a economizar tempo e dinheiro.

A **UPS** desenvolveu ferramentas de planejamento de rede chamado Network Planning Tools (NPT) para visualizar de forma abrangente as operaes da organizao. 

Essas ferramentas foram projetadas para fornecer uma viso geral das atividades nas instalaes da UPS em todo o mundo e do volume e distribuio de pacotes em sua rede de coleta e entrega. Tambm permite que a UPS obtenha informaes detalhadas sobre remessas em trnsito, como peso, volume e prazos de entrega.

A NPT utiliza aprendizado de mquina para **processar e analisar grandes quantidades de dados do mundo real, permitindo que a UPS visualize todas as suas instalaes, desvie seus pacotes durante tempestades ou mova rapidamente uma grande remessa.**

Assim como as melhores ferramentas de cincia de dados, o NPT capacita os humanos a tomar melhores decises. Quando um pacote  redirecionado, o aplicativo notifica um engenheiro no novo local sobre o plano revisado. O engenheiro ento avalia vrias opes e toma medidas, potencialmente redirecionando o pacote com base em novas informaes. 
O NPT tambm verifica a escolha do engenheiro para garantir que obteve o resultado desejado. Isso permite que o algoritmo do NPT melhore e tome decises mais inteligentes quando

Os motoristas da empresa fazem, em mdia, 120 entregas por dia.

**Ao cruzar os dados dos sensores nos veculos, requisitos das encomendas dos clientes e dados de mapas e geolocalizao, a UPS reduziu em 85 milhes o nmero de milhas percorridas por ano.**

Alm do aumento de qualidade de vida para os funcionrios, a medida proporcionou uma economia expressiva de combustvel.

## Caso de Uso 2: Bancos


A cincia de dados revolucionou a forma como as empresas **detectam atividades fraudulentas**. Ele desempenha um papel crucial na coleta, resumo e previso de bancos de dados de clientes para identificar atividades fraudulentas com preciso.

Inteligncia de Contrato (COiN)
O JP Morgan tambm usa aprendizado de mquina e inteligncia artificial para desenvolver um novo software chamado COIN (Contract Intelligence). A ferramenta **revisou milhares de contratos de crdito do banco**, utilizando reconhecimento de imagem para reconhecer padres nesses contratos. Esta tecnologia emprega aprendizagem no supervisionada, permitindo ao algoritmo classificar clusulas em um dos cerca de 150 atributos diferentes de contratos de crdito. Por exemplo, o algoritmo pode observar padres especficos com base no texto da clusula ou na localizao do acordo.

O investimento do JPMorgan em cincia de dados  motivado pelos benefcios da **reduo de custos**. O software pode revisar contratos em segundos, um processo que anteriormente exigia mais de 360 mil horas de trabalho dos advogados. O algoritmo no apenas economiza tempo e dinheiro. Tambm melhora a preciso do processo. Em comparao com advogados humanos, a preciso do COIN  significativamente maior. Assim, o investimento do banco no se resume apenas  reduo de custos, mas tambm ao **aumento da qualidade**.
